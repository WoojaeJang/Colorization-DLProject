# path 설정과 이미지 이름 확인

```python
import glob
import os
from os import listdir, makedirs
from os.path import isfile, join
```

```python
original_img_path = './data/resize_test/01_original'
resized_img_path = './data/resize_test/02_resized(512x512)'
```

```python
img_name_list = list(filter(lambda f: isfile(join(original_img_path, f)), listdir(original_img_path)))
```

```python
img_name_list
```

## 사이즈 변환해서 output에 저장

```python
import cv2
import time
from IPython.display import clear_output
```

```python
start = time.time()

i = 0
for img_name in img_name_list :
    
    clear_output(wait=True)
    
    
    try:
        original_img = cv2.imread(os.path.join(original_img_path, img_name))
        resized_img = cv2.resize(original_img, dsize = (512, 512), interpolation = cv2.INTER_CUBIC)
        cv2.imwrite(os.path.join(resized_img_path, img_name), resized_img)
    except:
        print('{}가 변환되지 않았습니다.'.format(img_name))
    
    
    i += 1
    print('진행 상황 : ', i,'/',len(img_name_list))
    print('진행 시간 : ', round((time.time() - start), 2))
    
print('총 걸린 시간 : ', round((time.time() - start),2), '초')
```

## input 사이즈 저장

```python
from PIL import Image
```

```python
original_img_size_list = []
for img_name in img_name_list:
    original_img = Image.open(os.path.join(original_img_path, img_name))
    original_img_size_list.append(original_img.size)

print(len(original_img_size_list))
print()

for original_img_size in original_img_size_list :
    print(original_img_size)
```

## 사이즈 되돌리기

```python
restored_img_path = './data/resize_test/03_restore'
```

```python
start = time.time()

i = 0
for img_name in img_name_list :
    
    clear_output(wait=True)
    
    
    try:
        resized_img = cv2.imread(os.path.join(resized_img_path, img_name))
        restored_img = cv2.resize(resized_img, dsize = original_img_size_list[i], interpolation = cv2.INTER_CUBIC)
        cv2.imwrite(os.path.join(restored_img_path, img_name), restored_img)
    except:
        print('{}가 변환되지 않았습니다.'.format(i))
        
    i += 1
    print('진행 상황 : ', i,'/',len(img_name_list))
    print('진행 시간 : ', round((time.time() - start), 2))
    
print('총 걸린 시간 : ', round((time.time() - start),2), '초')
```
